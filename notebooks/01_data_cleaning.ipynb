# Data Cleaning Project

### Dataset: Sample Sales Data

### Dataset Source: Kaggle Sample Sales Dataset

### Objective
The objective of this notebook is to prepare raw sales data for analysis and modeling by performing data cleaning tasks. The cleaning process includes:

- Inspecting dataset structure and summary statistics
- Identifying and handling missing values
- Checking and removing duplicate records
- Converting incorrect data types into appropriate formats
- Preparing a clean dataset for further exploratory data analysis (EDA)


## 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

## 2. Load Dataset

```python
df=pd.read_csv("sales_data_sample.csv",encoding="latin1")
df.head()
```

```python
df.columns
```

This dataset contains global sales transactions of collectible vehicle products, including customer details, product categories, pricing, and sales revenue.

## 4. Check Missing Values 

```python
df.isna().sum()
```

## 5. Handle Missing Values

```python
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna('Not Available')
```

```python
df['STATE'] = df['STATE'].fillna('Unknown')
```

```python
df['POSTALCODE'] = df['POSTALCODE'].fillna('Unknown')
```

```python
df['TERRITORY'] = df['TERRITORY'].fillna('Unknown')
```

## 6. Check Duplicate Rows

```python
df.duplicated().sum()
```

Duplicate records were checked using the duplicated() function.
No duplicate rows were found in the dataset.

## 7. Fix Data Types

```python
df['ORDERDATE']=pd.to_datetime(df['ORDERDATE'])
```

```python
df.info()
```

```python
df['ORDERDATE'].dt.year
```

```python
df['ORDERDATE'].dt.month
```

## 8. Check Data After Cleaning

```python
df.isna().sum()
```

## 9. Save Clean Dataset

```python
df.to_csv("cleaned_sales_data.csv", index=False)
```

## Conclusion

The dataset was successfully cleaned and prepared for further analysis. The following steps were performed:

- Missing values in ADDRESSLINE2, STATE, POSTALCODE, and TERRITORY were handled by replacing them with appropriate placeholder values.
- The dataset was checked for duplicate records, and no duplicates were found.
- The ORDERDATE column was converted into datetime format to enable time-based analysis.

The cleaned dataset is now ready for exploratory data analysis and machine learning tasks.
